```python
from selenium import webdriver
from time import sleep

driver = webdriver.Chrome('chromedriver.exe')
# 仙台のIT企業情報ページを開く
driver.get('http://www.misadb.jp/')
# キーワードを入力
elem_keyword = driver.find_element_by_id('keyword')
elem_keyword.send_keys('青葉区 WEB系')
# 検索
elem_btn = driver.find_element_by_xpath("/html/body/div[@id='wrap']/div[@id='mainBody']/div[@class='mainContent']/form[@id='form1']/div[@class='search']/dl/dd[@class='btn']/input")
elem_btn.click()
```

```python
import pandas as pd

companies = []
representatives = []
adresses = []
inquiries = []
establishments = []
capitals = []
revenues = []
personnels = []

page_lens = []
k = 1

# ページ数を取得
elems_page_len = driver.find_elements_by_xpath("/html/body/div[@id='wrap']/div[@id='mainBody']/div[@class='mainContent']/div[@class='listBox']/div[@class='pageList']/ul/li/a")
for elem_page_len in elems_page_len:
  page_len = elem_page_len.text
  page_lens.append(page_len)

# ページ数だけ繰り返し
while k < len(page_lens):
  #ページをクリック
  elem_page = driver.find_element_by_xpath("/html/body/div[@id='wrap']/div[@id='mainBody']/div[@class='mainContent']/div[@class='listBox']/div[@class='pageList']/ul/li["+str(k)+"]")
  elem_page.click()
  #該当ページの会社数取得用
  companies_search = [] 
  #社名取得
  elems_dt = driver.find_elements_by_tag_name('dt')
  for elem_dt in elems_dt:
    company = elem_dt.text
    companies.append(company)
    companies_search.append(company) #会社数
    sleep(1)

  i = 3
  #該当ページの会社数だけ繰り返し
  while i < len(companies_search)+3:
    # 会社情報ページをクリック
    elem_company_page = driver.find_element_by_xpath("/html/body/div[@id='wrap']/div[@id='mainBody']/div[@class='mainContent']/div[@class='listBox']/div["+str(i)+"]/div[@class='cont']/ul[@class='btn']/li[2]")
    elem_company_page.click()
    # 代表者名を取得
    elem_representative = driver.find_element_by_xpath("/html/body/div[@id='wrap']/div[@id='mainBody']/div[@class='mainContent']/div[@class='companyBox']/div[@id='tabvanilla']/div[@id='main1']/div[@class='company']/table[@class='fo14']/tbody/tr[1]/td")
    representative = elem_representative.text
    representatives.append(representative)
    # 住所を取得
    elem_adress = driver.find_element_by_xpath("/html/body/div[@id='wrap']/div[@id='mainBody']/div[@class='mainContent']/div[@class='companyBox']/div[@id='tabvanilla']/div[@id='main1']/div[@class='company']/table[@class='fo14']/tbody/tr[2]/td")
    adress = elem_adress.text
    adresses.append(adress)
    # 問い合わせを取得
    elem_inquiry = driver.find_element_by_xpath("/html/body/div[@id='wrap']/div[@id='mainBody']/div[@class='mainContent']/div[@class='companyBox']/div[@id='tabvanilla']/div[@id='main1']/div[@class='company']/table[@class='fo14']/tbody/tr[3]/td")
    inquiry = elem_inquiry.text
    inquiries.append(inquiry)
    # 設立を取得
    elem_establishment = driver.find_element_by_xpath("/html/body/div[@id='wrap']/div[@id='mainBody']/div[@class='mainContent']/div[@class='companyBox']/div[@id='tabvanilla']/div[@id='main1']/div[@class='company']/table[@class='fo14']/tbody/tr[4]/td")
    establishment = elem_establishment.text
    establishments.append(establishment)
    # 資本金を取得
    elem_capital = driver.find_element_by_xpath("/html/body/div[@id='wrap']/div[@id='mainBody']/div[@class='mainContent']/div[@class='companyBox']/div[@id='tabvanilla']/div[@id='main1']/div[@class='company']/table[@class='fo14']/tbody/tr[5]/td")
    capital = elem_capital.text
    capitals.append(capital)
    # 売上高を取得
    elem_revenue = driver.find_element_by_xpath("/html/body/div[@id='wrap']/div[@id='mainBody']/div[@class='mainContent']/div[@class='companyBox']/div[@id='tabvanilla']/div[@id='main1']/div[@class='company']/table[@class='fo14']/tbody/tr[6]/td")
    revenue = elem_revenue.text
    revenues.append(revenue)
    # 社員数を取得
    elem_personnel = driver.find_element_by_xpath("/html/body/div[@id='wrap']/div[@id='mainBody']/div[@class='mainContent']/div[@class='companyBox']/div[@id='tabvanilla']/div[@id='main1']/div[@class='company']/table[@class='fo14']/tbody/tr[7]/td")
    personnel = elem_personnel.text
    personnels.append(personnel)

    i = i + 1
    sleep(1)
    driver.back() # 戻る
  
  # 1ページ目のときは戻らない
  if k == 1:
    sleep(1)
    k = k + 1
  else:
    driver.back()
    sleep(1)
    k = k + 1 

# csvに出力
df = pd.DataFrame()
df['会社名'] = companies
df['代表者'] = representatives
df['住所'] = adresses
df['問い合わせ先'] = inquiries
df['設立年'] = establishments
df['資本金'] = capitals
df['売上高'] = revenues
df['社員数'] = personnels
df.to_csv('IT企業情報in青葉区前処理.csv', encoding='utf_8_sig', index=False)
#住所の建物の項目の前にスペースがなかったら追加
```

```python
# CSVファイル整形

# CSVファイルを開く
csv_file = open('IT企業情報in青葉区前処理.csv', encoding='utf_8_sig')
# 列を取得
ad_list = []
for row in csv.reader(csv_file):
    ad_list.append(row[2]) #取得したい列番号を指定（0始まり）
# 先頭行を削除しておく
del ad_list[0]

ad1_list = []
ad2_list = []
ad3_list = []
for ad in ad_list:
    ad_split = ad.split()
    if 2 == len(ad_split):
        ad1 = ad_split[0]
        ad2 = ad_split[1]
        ad3 = ''
        ad1_list.append(ad1)
        ad2_list.append(ad2)
        ad3_list.append(ad3)
    elif 3 == len(ad_split):
        ad1 = ad_split[0]
        ad2 = ad_split[1]
        ad3 = ad_split[2]
        ad1_list.append(ad1)
        ad2_list.append(ad2)
        ad3_list.append(ad3)

# csvに出力
df = pd.DataFrame()
df['会社名'] = companies
df['代表者'] = representatives
df['郵便番号'] = ad1_list
df['住所'] = ad2_list
df['建物名'] = ad3_list
df['問い合わせ先'] = inquiries
df['設立年'] = establishments
df['資本金'] = capitals
df['売上高'] = revenues
df['社員数'] = personnels
df.to_csv('IT企業情報in青葉区.csv', encoding='utf_8_sig', index=False)
```
